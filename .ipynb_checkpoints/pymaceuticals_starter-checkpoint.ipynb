## Observations and Insights 

Observations from data analysis

 Observation #1 - The population being split almost 50/50 between male and female 
                  may indicate that gender is not a factor in the effectiveness of the regimens.
                  

 Observation #2 - Both Capomulin and Ramicane had a higher level of testing conducted (i.e highest number of
                  mice in testing) and they had similar results in terms of effectiveness. This slightly elevated 
                  level of testing may account for some of the uniformity in the results 
                  observed as there is a larger population of results gathered.

 Observation #3 - Capomulin was highly effective in reducing the size of tumors over the time period under testing.

 Observation #4 - There may be a correlation between average mouse weight and average tumor size indicating that,
                  potentially, the larger the mouse, the heavier the tumor.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy import stats
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_results = pd.merge(left=mouse_metadata, right=study_results,how='inner', left_on='Mouse ID', right_on='Mouse ID')

# Display the data table for preview
combined_results
```

```python
# Checking the number of mice.
len(combined_results.index)
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
# Optional: Get all the data for the duplicate mouse ID. 
duplicated_entries = combined_results[combined_results.duplicated(['Mouse ID','Timepoint'], keep=False)]
duplicated_entries
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.

clean_data = combined_results.drop_duplicates(['Mouse ID','Timepoint'], keep = 'first')

clean_data
```

```python
# Checking the number of mice in the clean DataFrame.
len(clean_data.index)
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# This method produces everything in a single groupby function
drug_group = (clean_data.groupby("Drug Regimen")["Tumor Volume (mm3)"])
mouse_by_drug = (clean_data.groupby("Drug Regimen")["Mouse ID"])
mouse_by_drug = mouse_by_drug.agg(['count'])
drug_group_stats = drug_group.agg(['mean', 'median','var','std','sem'])
drug_group_stats
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
last_timepoint = clean_data.groupby('Mouse ID')['Timepoint'].max()

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
last_timepoint_join = pd.merge(left=clean_data, right=last_timepoint,how='inner', 
                               left_on=['Mouse ID','Timepoint'], right_on=['Mouse ID','Timepoint'])
last_timepoint_join
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drugs_list = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol = list()

# Calculate the IQR and quantitatively determine if there are any potential outliers.   
# Locate the rows which contain mice on each drug and get the tumor volumes
tumor_vol = last_timepoint_join[last_timepoint_join['Drug Regimen'].isin(drugs_list)]

# Check data extracted
tumor_vol.plot.bar(x='Mouse ID', y='Tumor Volume (mm3)', rot=90, figsize=(20,16))
```

```python
# add subset
capomulin_subset = tumor_vol[tumor_vol["Drug Regimen"] == "Capomulin"]
ramicane_subset = tumor_vol[tumor_vol["Drug Regimen"] == "Ramicane"]
infubinol_subset = tumor_vol[tumor_vol["Drug Regimen"]== "Infubinol"]
ceftamin_subset = tumor_vol[tumor_vol["Drug Regimen"]=="Ceftamin"]
```

```python
# Do quartile calculations for each drug
# Determine outliers using upper and lower bounds

#Campomulin
capomulin_quartiles = capomulin_subset["Tumor Volume (mm3)"].quantile([.25,.5,.75]).round(2)
capomulin_lowerq = capomulin_quartiles[0.25].round(2)
capomulin_upperq = capomulin_quartiles[0.75].round(2)
capomulin_iqr = round(capomulin_upperq-capomulin_lowerq,2)
capomulin_lower_bound = round(capomulin_lowerq - (1.5*capomulin_iqr),2)
capomulin_upper_bound = round(capomulin_upperq + (1.5*capomulin_iqr),2)

print(f"--------------------------------------------------------")
print(f"Capomulin IQR data is:")
print(f"The lower quartile of Capomulin is: {capomulin_lowerq}")
print(f"The upper quartile of Capomulin is: {capomulin_upperq}")
print(f"The interquartile range of Capomulin is: {capomulin_iqr}")
print(f"The the median of Capomulin is: {capomulin_quartiles[0.5]}")
print(f"Values below {capomulin_lower_bound} for Capomulin could be outliers.")
print(f"Values above {capomulin_upper_bound} for Capomulin could be outliers.")
print(f"--------------------------------------------------------")

#Ramicane
ramicane_quartiles = ramicane_subset["Tumor Volume (mm3)"].quantile([.25,.5,.75]).round(2)
ramicane_lowerq = ramicane_quartiles[0.25].round(2)
ramicane_upperq = ramicane_quartiles[0.75].round(2)
ramicane_iqr = round(ramicane_upperq-ramicane_lowerq,2)
ramicane_lower_bound = round(ramicane_lowerq - (1.5*ramicane_iqr),2)
ramicane_upper_bound = round(ramicane_upperq + (1.5*ramicane_iqr),2)

print(f"--------------------------------------------------------")
print(f"Ramicane IQR data is:")
print(f"The lower quartile of Ramicane is: {ramicane_lowerq}")
print(f"The upper quartile of Ramicane is: {ramicane_upperq}")
print(f"The interquartile range of Ramicane is: {ramicane_iqr}")
print(f"The the median of Ramicane is: {ramicane_quartiles[0.5]}")
print(f"Values below {ramicane_lower_bound} for Ramicane could be outliers.")
print(f"Values above {ramicane_upper_bound} for Ramicane could be outliers.")
print(f"--------------------------------------------------------")

#Infubinol
infubinol_quartiles = infubinol_subset["Tumor Volume (mm3)"].quantile([.25,.5,.75]).round(2)
infubinol_lowerq = infubinol_quartiles[0.25].round(2)
infubinol_upperq = infubinol_quartiles[0.75].round(2)
infubinol_iqr = round(infubinol_upperq-infubinol_lowerq,2)
infubinol_lower_bound = round(infubinol_lowerq - (1.5*infubinol_iqr),2)
infubinol_upper_bound = round(infubinol_upperq + (1.5*infubinol_iqr),2)

print(f"--------------------------------------------------------")
print(f"Infubinol IQR data is:")
print(f"The lower quartile of Infubinol is: {infubinol_lowerq}")
print(f"The upper quartile of CInfubinol is: {infubinol_upperq}")
print(f"The interquartile range of Infubinol is: {infubinol_iqr}")
print(f"The the median of Infubinol is: {infubinol_quartiles[0.5]}")
print(f"Values below {infubinol_lower_bound} for Infubinol could be outliers.")
print(f"Values above {infubinol_upper_bound} for Infubinol could be outliers.")
print(f"--------------------------------------------------------")

# Ceftamin
ceftamin_quartiles = ceftamin_subset["Tumor Volume (mm3)"].quantile([.25,.5,.75]).round(2)
ceftamin_lowerq = ceftamin_quartiles[0.25].round(2)
ceftamin_upperq = ceftamin_quartiles[0.75].round(2)
ceftamin_iqr = round(ceftamin_upperq-ceftamin_lowerq,2)
ceftamin_lower_bound = round(ceftamin_lowerq - (1.5*ceftamin_iqr),2)
ceftamin_upper_bound = round(ceftamin_upperq + (1.5*ceftamin_iqr),2)


print(f"--------------------------------------------------------")
print(f"Ceftamin IQR data is:")
print(f"The lower quartile of Ceftamin is: {ceftamin_lowerq}")
print(f"The upper quartile of Ceftamin is: {ceftamin_upperq}")
print(f"The interquartile range of Ceftamin is: {ceftamin_iqr}")
print(f"The the median of Ceftamin is: {ceftamin_quartiles[0.5]}")
print(f"Values below {ceftamin_lower_bound} for Ceftamin could be outliers.")
print(f"Values above {ceftamin_upper_bound} for Ceftamin could be outliers.")
print(f"--------------------------------------------------------")
```

## Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomuli
capomulin_data = clean_data.loc[clean_data['Drug Regimen'] == 'Capomulin']
cap_time_tumor = capomulin_data.groupby(['Timepoint']).agg('mean')

cap_time_plot = plt.plot(cap_time_tumor.index,cap_time_tumor['Tumor Volume (mm3)'], 
          color="blue",label="Time point vs tumor volume for Capomuli")
title_cap_plot = "Time point vs tumor volume for Capomulin regimen"
plt.title(title_cap_plot)
plt.suptitle('')
plt.xlabel("Time point")
plt.ylabel("Tumor Volume (mm3)")
plt.grid()
plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
cap_weight_tumor = capomulin_data.groupby(['Mouse ID']).agg(
                    mouse_weight=('Weight (g)', np.mean),
                    tumor_vol_mean=('Tumor Volume (mm3)', np.mean))

plt.scatter(cap_weight_tumor["mouse_weight"],cap_weight_tumor["tumor_vol_mean"])
plt.xlabel('Weight')
plt.ylabel('Tumor Volume (mm3)')
plt.title("Avg mouse weight vs avg tumor volume for Capomulin regimen")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

# Correlation coefficient using Pearson Correlation Coefficient 
correlation = st.pearsonr(cap_weight_tumor["mouse_weight"],cap_weight_tumor["tumor_vol_mean"])
print(f"The correlation between both factors is {round(correlation[0],2)}")
```

```python
# Perform a linear regression on weight vs tumor volume
wt_slope, wt_int, wt_r, wt_p, wt_std_err = stats.linregress(cap_weight_tumor["mouse_weight"],cap_weight_tumor["tumor_vol_mean"])
wt_fit = wt_slope * cap_weight_tumor["mouse_weight"] + wt_int
```

```python
plt.scatter(cap_weight_tumor["mouse_weight"],cap_weight_tumor["tumor_vol_mean"])
plt.plot(cap_weight_tumor["mouse_weight"],wt_fit,"--")
plt.xlabel('Weight')
plt.ylabel('Tumor Volume (mm3)')
plt.title("Avg mouse weight vs avg tumor volume for Capomulin regimen")
plt.show()
```
